```python
from matplotlib import pyplot as plt
import pandas as pd
from collections import Counter
```

```python
#Reading our data into dataframe
df = pd.read_csv('data.csv')
```

```python
#Creating objects from columns in our data
ids = df['Responder_id']
languages = df['LanguagesWorkedWith']
```

```python
#Splitting the ';' separated languages and taking count of each
#Final output is list of tuples containing language and its frequency
languages_f = Counter()
for row in languages:
    languages_f.update(row.split(';'))
    
```

```python
#Some useful methods built into counter objects
languages_f.most_common(15)
```

```python
#Separating the tuples into a list of languages and frequency
program_lang = []
popularity = []

for item in languages_f.most_common(15):
    program_lang.append(item[0])
    popularity.append(item[1])
    
print(program_lang, popularity)
```

```python
program_lang.reverse()
popularity.reverse()
```

```python
#print(plt.style.available)

plt.style.use('seaborn-whitegrid')

#plt.bar(program_lang, popularity)
plt.barh(program_lang, popularity)

plt.xlabel('Populairty')
plt.title('Popularity of different proramming languages')

plt.show()
```
